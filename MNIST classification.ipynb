## Convolutional Neural Networks(CNNs) architecture

````python
########################################################################################
### Place all code in this cell. You may also create new code cells below. (5 MARKS) ###
########################################################################################

from torch import nn
from torch.nn import functional as F

""" CNN model """
class CNNmodel(nn.Module):
    def __init__(self):
        # call the parent class
        super(CNNmodel, self).__init__()

        # first layer
        self.conv1 = nn.Conv2d(in_channels=1, out_channels=32, kernel_size=3, padding=1)
        self.bn1 = nn.BatchNorm2d(32)
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2)

        # second layer
        self.conv2 = nn.Conv2d(in_channels=32, out_channels=64, kernel_size=3, padding=1)
        self.bn2 = nn.BatchNorm2d(64)

        # third layer
        self.conv3 = nn.Conv2d(in_channels=64, out_channels=128, kernel_size=3, padding=1)
        self.bn3 = nn.BatchNorm2d(128)

        # fully connected layers
        self.fc1 = nn.Linear(128 * 3 * 3, 512)
        self.fc2 = nn.Linear(512, 256)
        self.fc3 = nn.Linear(256, 10)

        # dropout with 50% probability --> regularization
        self.dropout = nn.Dropout(0.5)

    # forward pass
    def forward(self, x):
        # apply conv layers, relu activation, batchNorm, and poolings
        x = self.pool(F.relu(self.bn1(self.conv1(x))))
        x = self.pool(F.relu(self.bn2(self.conv2(x))))
        x = self.pool(F.relu(self.bn3(self.conv3(x))))

        # flatten the output from conv layers
        x = x.view(-1, 128 * 3 * 3)

        # apply fully connected layers with relu
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))

        # dropout
        x = self.dropout(x)

        # fully connected --> output
        x = self.fc3(x)
        return x
````

````python
import torchvision
from torchvision import transforms
from torch.utils.data import DataLoader, random_split

# add data transformations with augmentation for performance enhancement
data_transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.1307,), (0.3081,))
])

# load MNIST dataset
trainset = torchvision.datasets.MNIST(root="./data", train=True, download=True, transform=data_transform)
testset = torchvision.datasets.MNIST(root="./data", train=False, download=True, transform=data_transform)

print("Full Training dataset size: {}".format(len(trainset)))
print("Testing dataset size: {}\n".format(len(testset)))

# split sizes
train_size = 50000
val_size = 10000
training, validation = random_split(trainset, [train_size, val_size])

# data loaders
# train_loader = DataLoader(training, batch_size=128, shuffle=True, num_workers=4, pin_memory=True)
val_loader = DataLoader(validation, batch_size=128, shuffle=False, num_workers=4, pin_memory=True)
test_loader = DataLoader(testset, batch_size=128, shuffle=False, num_workers=4, pin_memory=True)

# check the length after spliting
print(f"Training samples: {len(training)}")
print(f"Validation samples: {len(validation)}")
print(f"Test samples: {len(testset)}")
````

````python
import torch
from torch.cuda.amp import GradScaler, autocast

""" experiment 1 training function """
def experiment1_train_model(model, train_loader, val_loader, criterion, optimizer, epochs=50):
    # in case gpu is available
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model.to(device)

    # initialise gradiant
    scaler = GradScaler()

    # lists to store results
    experiment1_train_losses = []
    experiment1_val_losses = []
    experiment1_train_accuracies = []
    experiment1_val_accuracies = []

    # iterate over epochs
    for epoch in range(epochs):
        # set model to train mode
        model.train()

        # set default values
        running_train_losses = 0.0
        running_val_losses = 0.0
        correct_train = 0
        total_train = 0
        correct_val = 0
        total_val = 0

        # iterate over training data
        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.to(device)

            # zero param gradiants
            optimizer.zero_grad()

            # forward pass with mixed precision
            with autocast():
                outputs = model(inputs)
                loss = criterion(outputs, labels)

            # backward pass with gradiant scaling
            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()

            # update running training losses
            running_train_losses += loss.item()
            _, predicted = torch.max(outputs.data, 1)

            # update total and correct train counts
            total_train += labels.size(0)
            correct_train += (predicted == labels).sum().item()


        """ Validation """
        # set model to evaluation mode
        model.eval()
        # disable gradiant --> validation
        with torch.no_grad():

            # iterate validation data
            for inputs, labels in val_loader:
                inputs, labels = inputs.to(device), labels.to(device)

                # forward pass
                with autocast():
                    outputs = model(inputs)
                    loss = criterion(outputs, labels)

                # update values: running valid losses, total and correct counts
                running_val_losses += loss.item()
                # get predicted classes
                _, predicted = torch.max(outputs.data, 1)
                total_val += labels.size(0)
                correct_val += (predicted == labels).sum().item()

        # store and calculate average training and validation losses
        experiment1_train_losses.append(running_train_losses / len(train_loader))
        experiment1_val_losses.append(running_val_losses / len(val_loader))

        # store and calculate average trainign and validation accuracies
        experiment1_train_accuracies.append(100 * correct_train / total_train)
        experiment1_val_accuracies.append(100 * correct_val / total_val)

        print(f"Epoch {epoch+1}/{epochs}, Train Loss: {running_train_losses / len(train_loader):.4f}, Val Loss: {running_val_losses / len(val_loader):.4f}, Train Acc: {experiment1_train_accuracies[-1]:.2f}%, Val Acc: {experiment1_val_accuracies[-1]:.2f}%")

    # return records
    return experiment1_train_losses, experiment1_val_losses, experiment1_train_accuracies, experiment1_val_accuracies
````

````python
import torch
import torch.nn as nn
import torch.nn.functional as F


""" Underfitting """
class Underfitting_CNN(nn.Module):
    def __init__(self):
        super(Underfitting_CNN, self).__init__()

        # first layer
        self.conv1 = nn.Conv2d(in_channels=1, out_channels=32, kernel_size=3, padding=1)
        self.bn1 = nn.BatchNorm2d(32)
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2)

        # second layer
        self.conv2 = nn.Conv2d(in_channels=32, out_channels=64, kernel_size=3, padding=1)
        self.bn2 = nn.BatchNorm2d(64)

        # thrid layer
        self.conv3 = nn.Conv2d(in_channels=64, out_channels=128, kernel_size=3, padding=1)
        self.bn3 = nn.BatchNorm2d(128)

        # adjusted fully connected layers
        self.fc1 = nn.Linear(128 * 3 * 3, 32)
        self.fc2 = nn.Linear(32, 32)

        # output
        self.fc3 = nn.Linear(32, 10)

        # set dropout
        self.dropout = nn.Dropout(0.5)

    # forward pass
    def forward(self, x):

        # apply conv, batch norm, relu activation and pooling
        x = self.pool(F.relu(self.bn1(self.conv1(x))))
        x = self.pool(F.relu(self.bn2(self.conv2(x))))
        x = self.pool(F.relu(self.bn3(self.conv3(x))))

        # flatten output
        x = x.view(-1, 128 * 3 * 3)

        # apply fully connected with relu
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))

        # apply dropout
        x = self.dropout(x)

        # output
        x = self.fc3(x)
        return x


""" Overfitting """
class Overfitting_CNN(nn.Module):
    def __init__(self):
        super(Overfitting_CNN, self).__init__()
        self.conv1 = nn.Conv2d(in_channels=1, out_channels=32, kernel_size=3, padding=1)
        self.bn1 = nn.BatchNorm2d(32)
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2)
        self.conv2 = nn.Conv2d(in_channels=32, out_channels=64, kernel_size=3, padding=1)
        self.bn2 = nn.BatchNorm2d(64)
        self.conv3 = nn.Conv2d(in_channels=64, out_channels=128, kernel_size=3, padding=1)
        self.bn3 = nn.BatchNorm2d(128)


        self.fc1 = nn.Linear(128 * 3 * 3, 1024)
        self.fc2 = nn.Linear(1024, 1024)
        self.fc3 = nn.Linear(1024, 10)
        self.dropout = nn.Dropout(0.5)

    def forward(self, x):
        x = self.pool(F.relu(self.bn1(self.conv1(x))))
        x = self.pool(F.relu(self.bn2(self.conv2(x))))
        x = self.pool(F.relu(self.bn3(self.conv3(x))))
        x = x.view(-1, 128 * 3 * 3)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.dropout(x)
        x = self.fc3(x)
        return x


""" in between = balance """
class InBetween_CNN(nn.Module):
    def __init__(self):
        super(InBetween_CNN, self).__init__()
        self.conv1 = nn.Conv2d(in_channels=1, out_channels=32, kernel_size=3, padding=1)
        self.bn1 = nn.BatchNorm2d(32)
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2)
        self.conv2 = nn.Conv2d(in_channels=32, out_channels=64, kernel_size=3, padding=1)
        self.bn2 = nn.BatchNorm2d(64)
        self.conv3 = nn.Conv2d(in_channels=64, out_channels=128, kernel_size=3, padding=1)
        self.bn3 = nn.BatchNorm2d(128)


        self.fc1 = nn.Linear(128 * 3 * 3, 256)
        self.fc2 = nn.Linear(256, 256)
        self.fc3 = nn.Linear(256, 10)
        self.dropout = nn.Dropout(0.5)

    def forward(self, x):
        x = self.pool(F.relu(self.bn1(self.conv1(x))))
        x = self.pool(F.relu(self.bn2(self.conv2(x))))
        x = self.pool(F.relu(self.bn3(self.conv3(x))))
        x = x.view(-1, 128 * 3 * 3)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.dropout(x)
        x = self.fc3(x)
        return x
````

````python
import torch
from torchvision import transforms, datasets
from torch.utils.data import DataLoader, random_split, ConcatDataset

""" original and augmented transformation (translation only) """
original_transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.1307,), (0.3081,))
])

augment_transform = transforms.Compose([
    transforms.RandomAffine(degrees=0, translate=(0.2, 0.2)),
    transforms.ToTensor(),
    transforms.Normalize((0.1307,), (0.3081,))
])


"""
    original_train_dataset: load dataset of training with original transformation,
    augmented_datasets: load dataset of training with augment transformation multiple 3

"""
original_train_dataset = datasets.MNIST(root="./data", train=True, download=True, transform=original_transform)


"""
    split dataset into training and validation:
    orig_train, orig_val_dataset: 50,000 of training and 10,000 of validation,
    augmented_train_dataset: 50,000 * 4 of validation

"""
train_size = 50000
val_size = 10000
orig_train_dataset, orig_val_dataset = random_split(original_train_dataset, [train_size, val_size])

# iterate 3 times
augmented_train_dataset = []
for _ in range(3):
    augmented_datasets = datasets.MNIST(root="./data", train=True, download=True, transform=augment_transform)
    augment_train, _ = random_split(augmented_datasets, [train_size, val_size])
    augmented_train_dataset.append(augment_train)


# combine the orig_train_dataset with augmented_train_dataset
combined_train_dataset = ConcatDataset([orig_train_dataset] + augmented_train_dataset)

# data loaders
experiment2_train_loader = DataLoader(combined_train_dataset, batch_size=128, shuffle=True, num_workers=4, pin_memory=True)
experiment2_val_loader = DataLoader(orig_val_dataset, batch_size=128, shuffle=False, num_workers=4, pin_memory=True)

# load the original test dataset
experiment2_test_dataset = datasets.MNIST(root="./data", train=False, download=True, transform=original_transform)
experiment2_test_loader = DataLoader(experiment2_test_dataset, batch_size=128, shuffle=False, num_workers=4, pin_memory=True)

# ensure the accurate length of each data
print(f"Original train dataset length: {len(original_train_dataset)}")
print(f"Combined train dataset length: {len(combined_train_dataset)}")
print(f"Training set length: {len(combined_train_dataset)}")
print(f"Validation set length: {len(orig_val_dataset)}")
print(f"Test set length: {len(experiment2_test_dataset)}")
````

````python
import torch
from torch.cuda.amp import GradScaler, autocast

def experiment2_train_model(model, train_loader, val_loader, criterion, optimizer, epochs=50):
    # initialise gradiant scaler
    scaler = GradScaler()

    # lists for results
    train_losses = []
    val_losses = []
    train_accuracies = []
    val_accuracies = []

    # iterate over each epoch
    for epoch in range(epochs):
        # set model to train mode
        model.train()

        # set default values
        running_train_loss = 0.0
        correct_train = 0
        total_train = 0

        # iterate over training data
        for inputs, labels in train_loader:
            # zero params gradiant
            optimizer.zero_grad()

            # forward pass with gradiant scaler
            with autocast():
                outputs = model(inputs)
                loss = criterion(outputs, labels)

            # backward pass with gradiant scaler
            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()

            # get predicted
            _, preds = torch.max(outputs, 1)

            # update running training loss, correct counts, total count
            running_train_loss += loss.item() * inputs.size(0)
            correct_train += (preds == labels).sum().item()
            total_train += labels.size(0)

        # store and calculate average of both training and accuracy losses for each epoch
        epoch_train_loss = running_train_loss / total_train
        epoch_train_acc = correct_train / total_train
        train_losses.append(epoch_train_loss)
        train_accuracies.append(epoch_train_acc)

        """ Validation """
        # set model to evaluation mode
        model.eval()

        # set default values
        running_val_loss = 0.0
        correct_val = 0
        total_val = 0

        # disable gradiant calculation
        with torch.no_grad():
            for inputs, labels in val_loader:

                # forward pass
                outputs = model(inputs)
                loss = criterion(outputs, labels)

                # get prediction
                _, preds = torch.max(outputs, 1)

                # update values
                running_val_loss += loss.item() * inputs.size(0)
                correct_val += (preds == labels).sum().item()
                total_val += labels.size(0)

        # store and calculate losses and accuracies
        epoch_val_loss = running_val_loss / total_val
        epoch_val_acc = correct_val / total_val
        val_losses.append(epoch_val_loss)
        val_accuracies.append(epoch_val_acc)

        print(f'Epoch {epoch+1}/{epochs}, Train Loss: {epoch_train_loss:.4f}, Train Acc: {epoch_train_acc:.4f}, Val Loss: {epoch_val_loss:.4f}, Val Acc: {epoch_val_acc:.4f}')

    # return records
    return train_losses, val_losses, train_accuracies, val_accuracies
````
